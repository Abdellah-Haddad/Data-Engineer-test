# **Extraction de données**

Importation des librairies nécessaires

```python
import requests
from bs4 import BeautifulSoup

from flask import Flask, jsonify
```

Dans un premier temps, nous définirons une méthode pour extraire les données nécessaires.




```python
#Methode pour l'extraction des donneés

def data_extractor(url):
  response = requests.get(url)

  if response.status_code == 200:
     soup = BeautifulSoup(response.content, 'html.parser')

     return soup

  else:
    print(f"Impossible d'accéder au lien {url}")
```

```python
# Extraction des données de chaque lien

urls = [
    "https://www.creditmutuel.fr/fr/particuliers/epargne/livret-de-developpement-durable.html",
    "https://www.monabanq.com/fr/produits-bancaires/livret-developpement-durable/en-resume.html",
    "https://www.banquepopulaire.fr/bpaura/epargner/livret-transition-energetique/"
]

extracted_data = [data_extractor(url) for url in urls]
print(extracted_data)
```

# **Création d'APIs/Web Services**

```python
app = Flask(__name__)

# variable pour stocker Les données extraites
data = extracted_data[0]

@app.route('/api/data', methods=['GET'])
def get_data():
    return jsonify(data)

if __name__ == '__main__':
    app.run(debug=True)
```
